# Editing Annotations

```python
# Create a new collection containing two objects 
import pycollector.backend
api = pycollector.backend.API()
api.new_collection(
    name='Orbit around a coffee cup and sunglasses',
    project_name="Editing Anotations",
    collection_description='Collect a video where you orbit the camera around a coffee cup and sunglasses',
    activities='mug_orbits',
    activity_short_names="orbit",
    actor='Mug',
    objects='Sunglasses'
)
```

```python
# Go to the app and record an annotated video for this new collection
v = pycollector.video.last()
```

```python
# Show the annotated first frame of this video
v[0].show()  
```

```python
# Go back to the app and edit the video to add the secondary object box and correct the primary object box, then fetch it again
v = pycollector.video.last()
v.frame(0).show()
```

```python
# Visualize frame 90 of the video showing the beginning of the camera orbit
v[90].show()
```
